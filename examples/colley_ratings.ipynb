```python
import penaltyblog as pb
```

# Load the Data

Load the fixtures and results from [football-data.co.uk](https://football-data.co.uk/).

```python
fbd = pb.scrapers.FootballData("ENG Premier League", "2023-2024")
df = fbd.get_fixtures()

df.head()
```

# Colley Ratings Including Tied Scorelines

```python
colley = pb.ratings.Colley(
    df["goals_home"],
    df["goals_away"],
    df["team_home"],
    df["team_away"]
)
ratings = colley.get_ratings()
print(ratings.head(10))
```

# Colley Ratings Excluding Tied Scorelines

You can exclude draws from the calculation by setting ``include_draws=False``. This focuses purely on decisive results:

```python
colley = pb.ratings.Colley(
    df["goals_home"],
    df["goals_away"],
    df["team_home"],
    df["team_away"],
    include_draws=False,
)
ratings = colley.get_ratings()
print(ratings.head(10))
```
